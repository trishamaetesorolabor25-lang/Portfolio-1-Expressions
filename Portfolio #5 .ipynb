```python
# Word Guessing Game

def word_game():
    secret_word = "chemeng"
    guessed_letters = "" # String to store all user guesses
    attempts = 6

    print("Welcome to the Word Guessing Game! ")

    while attempts > 0:
        # Display the current progress
        display_word = ""
        
        # Pattern: Looping through strings 
        for letter in secret_word:
            # Pattern: Using 'in' as a logical operator 
            if letter in guessed_letters:
                display_word = display_word + letter # Concatenation 
            else:
                display_word = display_word + "_" 
        
        print(f"\nWord: {display_word}")
        print(f"Attempts remaining: {attempts}")

        # Check if user has won
        if "_" not in display_word:
            print("Congratulations! You found the word!")
            break

        # Get user guess
        guess = input("Guess a letter: ").lower() 

        # Basic validation using 'in'
        if guess in guessed_letters:
            print("You already guessed that letter!")
            continue # Chapter 5: Iterations pattern

        guessed_letters = guessed_letters + guess

        # Logic to decrease attempts
        if guess not in secret_word:
            attempts = attempts - 1
            print(f"No '{guess}' in the word.")

    if attempts == 0:
        print(f"\nGame Over! The word was: {secret_word}")

word_game()
```

```python
# Word Puzzle Generator (Scrambler)

import random

def word_puzzle_generator():
    original_word = input("Enter a word to turn into a puzzle: ")
    
    # Standardize to lower case 
    word = original_word.lower()
    
    # Convert string to a list to allow shuffling 
    char_list = list(word)
    random.shuffle(char_list)
    
    # Reconstruct the scrambled string using a loop and concatenation
    scrambled = ""
    for char in char_list:
        scrambled = scrambled + char # Concatenation 
        
    print(f"\nOriginal Word: {original_word}")
    print(f"Scrambled Puzzle: {scrambled}")
    
    # User tries to solve it
    guess = input("\nCan you solve the puzzle? ").lower()
    
    # String Comparison 
    if guess == word:
        print("Correct! You solved the puzzle.")
    else:
        print(f"Not quite. The answer was {word}.")

word_puzzle_generator()
```

```python
# Autocomplete System

def autocomplete_system():
    # Your specific word list
    suggestions = [
        "chemical engineering", 
        "chemistry", 
        "calculus", 
        "computer science",  
        "physics",
        "lecture",
        "Philippines",
        "Manila",
        "Thomas"
    ]
    
    print("Autocomplete Tool")
    user_query = input("Start typing: ")
    
    # This ensures "manila" matches "Manila"
    search_term = user_query.lower()
    
    print("\nSuggestions:")
    found = False
    
    #Iterate through the list 
    for word in suggestions:
        # Check for prefix match 
        # We convert 'word' to lower temporarily for a fair comparison
        if word.lower().startswith(search_term):
            print(f" {word}")
            found = True
            
    # Fallback logic using 'in' operator 
    if not found:
        print("(No direct match. Checking keywords...)")
        for word in suggestions:
            if search_term in word.lower():
                print(f"  {word}")
                found = True

    if not found:
        print("No results found.")


autocomplete_system()
```
