```python
# Quiz in Chemical Engineering

# Void Function
def show_welcome():
    print("CHEMICAL ENGINEERING QUIZ")

# Fruitful Function
def check_answer(answer, correct):
    if answer == correct:
        return 1
    else:
        return 0

# Void Function
def show_result(score):

    print("\nRESULT")
    print("Score:", score, "/ 5")

    percentage = float(score) / 5 * 100

    print("Percentage:", percentage)

    if percentage >= 90:
        print("Rating: Excellent")

    elif percentage >= 75:
        print("Rating: Good")

    elif percentage >= 60:
        print("Rating: Passed")

    else:
        print("Rating: Failed")

# Main Program
def main():

    show_welcome()

    score = 0

    # Question 1
    print("\n1. What is the SI unit of pressure?")
    print("A. Joule")
    print("B. Watt")
    print("C. Pascal")
    print("D. Newton")

    answer1 = input("Your answer: ")

    result = check_answer(answer1, "C")

    if result == 1:
        score = score + 1
    else:
        print("Wrong in Question 1")
        print("Correct Answer: C")

    # Question 2
    print("\n2. What is the formula of water?")
    print("A. CO2")
    print("B. H2O")
    print("C. O2")
    print("D. NaCl")

    answer2 = input("Your answer: ")

    result = check_answer(answer2, "B")

    if result == 1:
        score = score + 1
    else:
        print("Wrong in Question 2")
        print("Correct Answer: B")

    # Question 3
    print("\n3. What process separates mixtures by boiling point?")
    print("A. Filtration")
    print("B. Distillation")
    print("C. Screening")
    print("D. Sedimentation")

    answer3 = input("Your answer: ")

    result = check_answer(answer3, "B")

    if result == 1:
        score = score + 1
    else:
        print("Wrong in Question 3")
        print("Correct Answer: B")

    # Question 4
    print("\n4. What does pH measure?")
    print("A. Pressure")
    print("B. Temperature")
    print("C. Density")
    print("D. Acidity or alkalinity")

    answer4 = input("Your answer: ")

    result = check_answer(answer4, "D")

    if result == 1:
        score = score + 1
    else:
        print("Wrong in Question 4")
        print("Correct Answer: D")

    # Question 5
    print("\n5. What is the unit of concentration?")
    print("A. Molarity")
    print("B. Energy")
    print("C. Velocity")
    print("D. Pressure")

    answer5 = input("Your answer: ")

    result = check_answer(answer5, "A")

    if result == 1:
        score = score + 1
    else:
        print("Wrong in Question 5")
        print("Correct Answer: A")

    show_result(score)

main()
```

```python
#BILL SPLITTER SYSTEM

#Void Function
def show_intro():
    print("BILL SPLITTER SYSTEM")

#Fruitful Function
def compute_total(bill, tip):

    total_tip = bill * (tip / 100)

    total = bill + total_tip

    return total

#Fruitful Function
def split_bill(total, people):

    payment = total / people

    return payment

#Void Function
def show_result(total, payment, people):

    print("\nRESULT")
    print("Total Bill:", total)
    print("Number of People:", people)
    print("Payment Per Person:", payment)

#Main Program
def main():

    show_intro()

    # Inputs
    bill = float(input("Enter total bill amount: "))
    people = int(input("Enter number of people: "))
    tip = float(input("Enter tip percentage: "))

    # Validation
    if people <= 0:
        print("Invalid number of people")

    else:

        # Computation
        total = compute_total(bill, tip)

        payment = split_bill(total, people)

        # Output
        show_result(total, payment, people)

# Run Program
main()
```

```python
#E-WALLET SYSTEM

#Users
users = {
    "Trisha": {
        "pin": "6969",
        "balance": 10000.0
    },

    "Thomas": {
        "pin": "6767",
        "balance": 5000.0
    }
}

#Void Function
def show_menu():

    print("\nE-WALLET MENU")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Transfer")
    print("5. Exit")

#Void Function
def receipt(action, amount):

    print("\nRECEIPT")
    print("Action:", action)
    print("Amount:", amount)

#Fruitful Function
def login(username, pin):

    if username in users:

        if pin == users[username]["pin"]:
            return 1

        else:
            return 0

    else:
        return 0

#Fruitful Function
def deposit(balance, amount):

    new_balance = balance + amount

    return new_balance

#Fruitful Function
def withdraw(balance, amount):

    if amount <= balance:

        new_balance = balance - amount

        return new_balance

    else:
        return balance

#Main Program
def main():

    username = input("Enter username: ")
    pin = input("Enter PIN: ")

    result = login(username, pin)

    if result == 1:

        print("Login Successful")

        while True:

            show_menu()

            choice = input("Choose option: ")

            # Check Balance
            if choice == "1":

                print("Balance:", users[username]["balance"])

            # Deposit
            elif choice == "2":

                amount = float(input("Enter amount: "))

                balance = users[username]["balance"]

                users[username]["balance"] = deposit(balance, amount)

                receipt("Deposit", amount)

            # Withdraw
            elif choice == "3":

                amount = float(input("Enter amount: "))

                balance = users[username]["balance"]

                if amount <= balance:

                    users[username]["balance"] = withdraw(balance, amount)

                    receipt("Withdraw", amount)

                else:
                    print("Insufficient Balance")

            # Transfer
            elif choice == "4":

                receiver = input("Enter receiver username: ")

                if receiver in users:

                    amount = float(input("Enter amount: "))

                    balance = users[username]["balance"]

                    if amount <= balance:

                        users[username]["balance"] = balance - amount

                        users[receiver]["balance"] = users[receiver]["balance"] + amount

                        print("Transfer Successful")

                    else:
                        print("Insufficient Balance")

                else:
                    print("User not found")

            # Exit
            elif choice == "5":

                print("Thank you for using E-Wallet")
                break

            else:
                print("Invalid Option")

    else:
        print("Invalid Login")

#Run Program
main()
```
