```python
# Multi-Subject Grade Tracker

def grade_tracker():
    # Initialize a list of lists (A list containing other lists) 
    subjects = [
        [85, 92, 78],  # Subject 1: Math
        [88, 90, 95],  # Subject 2: Science
        [70, 75, 80]   # Subject 3: History
    ]
    
    subject_names = ["Math", "Science", "History"]

    print("--- Student Grade Report ---")
    
    # definite loop with range(len()) to track indices 
    for i in range(len(subjects)):
        # Access the specific sub-list using an index 
        current_grades = subjects[i]
        name = subject_names[i]
        
        # built-in functions to analyze the sub-list 
        total = sum(current_grades)
        count = len(current_grades)
        average = total / count
        highest = max(current_grades)
        
        print(f"\nSubject: {name}")
        print(f"  Grades: {current_grades}")
        print(f"  Average: {average:.2f}")
        print(f"  Highest Grade: {highest}")

grade_tracker()
```

```python
# Guest List Manager

def manage_guests():
    # Start with an empty list (Building a list from scratch)
    guest_list = list() 
    
    print("--- Event Guest List Manager ---")
    print("Type 'done' when you are finished adding names.")

    # Indefinite Iteration: Keep asking until the user is 'done'
    while True: 
        name = input("Enter guest name: ").strip().title()
        
        if name.lower() == 'done':
            break 
            
        # Logical Operator: Check if the item is NOT in the list
        # This prevents duplicate entries
        if name not in guest_list: 
            # List Method: Add the element to the end of the list
            guest_list.append(name) 
            print(f"{name} has been added to the list.")
        else:
            # Result if the name is already a collection element
            print(f"Notice: {name} is already on the guest list.")

    # List Method: Sort the collection alphabetically
    guest_list.sort() 
    
    # Built-in Function: Count the elements
    total_guests = len(guest_list) 
    
    print("\n--- Final Sorted Guest List ---")
    print(f"Total Guests: {total_guests}")
    
    # Definite Iteration: Print each name in the final collection
    for guest in guest_list: 
        print(f" • {guest}")

manage_guests()
```

```python
# Grocery List "In-Stock" Checker

def pantry_checker():
    # Our current stock (A List Collection)
    pantry = ['egg', 'milk', 'sugar', 'flour', 'butter']
    
    print("--- Welcome to the Pantry Checker ---")
    print(f"Current Inventory: {pantry}")
    
    # Ask the user what they want to check
    search_item = input("\nWhat ingredient are you looking for? ").strip().lower()
    
    # Use the 'in' operator to check existence 
    if search_item in pantry:
        print(f"Yes! We have '{search_item}' in stock.")
    else:
        print(f"No, '{search_item}' is not in the pantry.")
        
        # Use 'append' to add the missing item (
        choice = input(f"Would you like to add {search_item} to the list? (yes/no): ")
        if choice.lower() == 'yes':
            pantry.append(search_item)
            # Sort the list to keep it organized 
            pantry.sort()
            print(f"Updated Pantry (Sorted): {pantry}")

pantry_checker()
```
