### <b>Student Grade Assessor<b>

<div>

```python
# Student Grade Assessor
# First Year Chemical Engineering

try:

    # User will input grades 
    chem_lect = float(input("Analytical Chemistry (Lecture): "))
    chem_lab = float(input("Analytical Chemistry (Laboratory): "))
    calc2 = float(input("Calculus 2: "))
    comp_fund = float(input("Computer Fundamentals and Programming: "))
    phys_lect = float(input("Physics for Engineers (Lecture): "))
    phys_lab = float(input("Physics for Engineers (Laboratory): "))
    fitness = float(input("Exercise-Based Fitness Activities: "))
    entrepreneur = float(input("The Entrepreneurial Mind: "))

    # Confirmation 
    grades = [chem_lect, chem_lab, calc2, comp_fund,
              phys_lect, phys_lab, fitness, entrepreneur]

    if any(g < 1 or g > 5 for g in grades):
        print("\nInvalid input! Grades must be between 1.0 and 5.0 only.")

    else:

        print("\nSUBJECT RESULTS ")

        subjects = [
            "Analytical Chemistry (Lecture)",
            "Analytical Chemistry (Laboratory)",
            "Calculus 2",
            "Computer Fundamentals and Programming",
            "Physics for Engineers (Lecture)",
            "Physics for Engineers (Laboratory)",
            "Exercise-Based Fitness Activities",
            "The Entrepreneurial Mind"
        ]

        for subject, grade in zip(subjects, grades):
            if grade <= 3.0:
                print(f"{subject}: PASSED")
            else:
                print(f"{subject}: FAILED")

        # Compute GWA
        gwa = sum(grades) / len(grades)

        print("\nSUMMARY ")
        print("GWA:", round(gwa, 2))

        # OVERALL STATUS
        if gwa <= 3.0:
            print("Overall Status: PASSED")
        else:
            print("Overall Status: FAILED")

        # HONORS
        print("\nHONORS ")

        # Check if all subjects are eligible (no grade above 2.44)
        eligible = all(g <= 2.44 for g in grades)

        if gwa >= 1.00 and gwa <= 1.44:
            if eligible:
                print("Honor: PRESIDENT'S LISTER")
            else:
                print("Honor: NOT QUALIFIED (subject grade above 2.44)")

        elif gwa <= 1.74:
            if eligible:
                print("Honor: DEAN'S LISTER")
            else:
                print("Honor: NOT QUALIFIED (subject grade above 2.44)")

        else:
            print("Honor: NONE")

except ValueError:
    print("Invalid input! Please enter numeric values only.")
```

### <b>Rock Paper Scissors Game<b>

<div>

```python
# Rock Paper Scissors Game
# Player vs Computer


import random

try:
    print("ROCK PAPER SCISSORS")
    print("Choices: rock, paper, scissors\n")

    # User input
    player = input("Enter your choice: ").lower()

    # Computer choice
    choices = ["rock", "paper", "scissors"]
    computer = random.choice(choices)

    print("\nComputer chose:", computer)

    # Validate input
    if player not in choices:
        print("Invalid choice! Please enter rock, paper, or scissors.")

    else:

        # Tie condition
        if player == computer:
            print("Result: It's a TIE!")

        # Player chooses rock
        elif player == "rock":
            if computer == "scissors":
                print("Result: YOU WIN!")
            else:
                print("Result: YOU LOSE!")

        # Player chooses paper
        elif player == "paper":
            if computer == "rock":
                print("Result: YOU WIN!")
            else:
                print("Result: YOU LOSE!")

        # Player chooses scissors
        elif player == "scissors":
            if computer == "paper":
                print("Result: YOU WIN!")
            else:
                print("Result: YOU LOSE!")

except Exception:
    print("Something went wrong. Please try again.")
```

### <b>Password Generator<b>

<div>

```python
# Password Generator
# This program generates a random password


import random

try:
    print("PASSWORD GENERATOR")

    # Ask password length
    length = int(input("Enter password length (e.g., 6-20): "))

    if length <= 0:
        print("Invalid length! Must be greater than 0.")

    else:

        # Character sets
        letters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
        numbers = "0123456789"
        symbols = "!@#$%^&*()_+-=[]{}|;:,.<>?/"

        all_chars = letters + numbers + symbols

        password = ""

        # Generate password
        for i in range(length):
            password += random.choice(all_chars)

        print("\nGenerated Password:", password)

except ValueError:
    print("Invalid input! Please enter a number only.")
```
