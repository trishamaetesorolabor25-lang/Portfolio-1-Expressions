```python
# Password Attempt System

password = "longlivechemeng"
attempts = 0

while True:
    # Taking input from the user
    guess = input('Enter your password: ')
    
    # Incrementing a counter variable 
    # Starts at 0 and adds one each time 
    attempts = attempts + 1
    
    # Logic to check the guess (
    if guess == password:
        print('Access Granted!')
        # The break statement ends the loop 
        break
    
    # Limiting attempts (Iteration control)
    if attempts >= 3:
        print('Too many incorrect attempts. Account locked.')
        break
    
    print('Incorrect password. Try again.')

# Statement immediately following the loop 
print('Done!')
```

```python
# Student Attendance Counter

count = 0

while True:  # Runs until a logical condition is met or break is called 
    name = input('Enter student name (type "done" to finish): ')
    
    if name == 'done':
        break  # Ends the current loop and jumps to the next statement 
        
    count = count + 1
    print('Added:', name, '| Current Count:', count)

print('Final Attendance Count:', count)
```

```python
# 7-Day Goal Tracker

daily_goal = 10000
weekly_goal = daily_goal * 7
total_steps = 0
days_met_goal = 0 # Counter for successful days

print(f"7-Day Activity Tracker (Goal: {daily_goal} steps/day) ")

# Using range(1, 8) to loop exactly 7 times (Day 1 through Day 7)
for day in range(1, 8):
    while True: # Nested loop to handle invalid input
        try:
            steps = int(input(f"Enter steps for Day {day}: "))
            break # Exit the nested loop if input is a valid number
        except ValueError:
            print("Invalid input. Please enter a whole number.")

    # Summing the total
    total_steps = total_steps + steps
    
    # Conditional logic to check daily goal
    if steps >= daily_goal:
        print(f"Day {day}: Goal Met! ")
        days_met_goal = days_met_goal + 1
    else:
        print(f"Day {day}: Under goal by {daily_goal - steps} steps.")

# Final Results After the Loop
print("\nWeekly Summary ")
print(f"Total Weekly Steps: {total_steps} / {weekly_goal}")
print(f"Days you hit the 10,000 step goal: {days_met_goal} out of 7")

if total_steps >= weekly_goal:
    print("Incredible! You hit your overall weekly target.")
else:
    print(f"You were {weekly_goal - total_steps} steps away from your weekly target.")
```
